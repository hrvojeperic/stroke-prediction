```python
# TASK 3
# Data Preparation. Detect and correct data quality problems (missing data, noise,
# outliers, etc.) and transform the data into an appropriate format for data analysis.
# Describe your data preparation process and report the results obtained.
```

```python
# drop columns that are uneeded. Such as id, work_type, and Residence_type
clean_healthcare_data = healthcare_data.copy(deep=True)
clean_healthcare_data.drop(columns=['id', 'work_type', 'Residence_type'], axis=1, inplace=True)
```

```python
# change gender using dummy variables. Male = 1, Female = 0
gender_column = numpy.where(clean_healthcare_data["gender"] == "Male", 1, 0)
# add binary data to set
clean_healthcare_data.drop(columns=['gender'], axis=1, inplace=True)
clean_healthcare_data["gender"] = gender_column
```

```python
# change ever_married using dummy variables. Married = 1, Never Married = 0
married_column = numpy.where(clean_healthcare_data["ever_married"] == "Yes", 1, 0)
# add binary data to set
clean_healthcare_data.drop(columns=['ever_married'], axis=1, inplace=True)
clean_healthcare_data["married"] = married_column
```

```python
# "formerly smoked", "never smoked", "smokes" or "Unknown"

# change "formerly smoked" using dummy variables. formerly smoked = 1, Never formerly smoked = 0
formerly_smoked_column = numpy.where(clean_healthcare_data["smoking_status"] == "formerly smoked", 1, 0)
# change "never_smoked" using dummy variables. never_smoked = 1, smoked = 0
never_smoked_column = numpy.where(clean_healthcare_data["smoking_status"] == "never smoked", 1, 0)
# change "formerly smoked" using dummy variables. smokes = 1, doesnt smokes = 0
smokes_column = numpy.where(clean_healthcare_data["smoking_status"] == "smokes", 1, 0)

# add binary data to set
clean_healthcare_data.drop(columns=['smoking_status'], axis=1, inplace=True)
clean_healthcare_data["formerly_smoked_yes"] = formerly_smoked_column
clean_healthcare_data["never_smoked_yes"] = never_smoked_column
clean_healthcare_data["smokes_yes"] = smokes_column
```

```python
# bmi contains null values for about 100 entries
# we decided to remove the entries since bmi is an important data point in our model and the number of null entries is
# relatively small.
clean_healthcare_data = clean_healthcare_data.dropna()
```

```python
# TASK 5 
# Data Modeling. Train and test models using the data. Your data modeling step must
# include at least two of the following tasks: (1) Regression, (2) Classification,
# (3) Clustering, (4) Text analysis, and (5) Time series analysis. Consider multiple 
# techniques, parameters, and variables. Describe your data modeling process and report
# the results obtained.
```

```python
clean_healthcare_data = clean_healthcare_data[["age","hypertension","heart_disease","avg_glucose_level","bmi","gender","married","formerly_smoked_yes","never_smoked_yes","smokes_yes", "stroke"]]

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X, y = smote.fit_resample(clean_healthcare_data.drop(['stroke'], axis=1), clean_healthcare_data[['stroke']])

smote_data = X;
smote_data["stroke"] = y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.70,
                                                    test_size=0.30,
                                                    random_state=0)
scaler = StandardScaler()
scaler.fit(X = x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# (2) Classification: Decision Trees & K-nearest
# Variable Groups:
# 1. Age and avg_glucose_level (best indicators) [0, 3]
# 2. Married, Gender, Age (most interesting indicator)  [0, 5, 6]
# 3. Currently smoking, hypertension, heart_disease (random indicators) [1, 2, 9]
# 4. Heart disease and BMI (Our best guess indicators)  [2, 4]
# 5. All variables [0, 1, 3, 4, 5, 6, 7, 8, 9]
```

```python
# Decision Trees
```

```python
# 1. Age and avg_glucose_level (best indicators)
classifier = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
classifier.fit(x_train_scaled[:,[0, 3]], y_train['stroke'])

y_pred = classifier.predict(x_test_scaled[:,[0, 3]])
conf_matrix = metrics.confusion_matrix(y_test['stroke'], y_pred)
sns.heatmap(conf_matrix, annot = True, fmt = ".3f", square = True, cmap = plt.cm.Blues)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion matrix')
plt.tight_layout()
```

```python
accuracy = metrics.accuracy_score(y_test['stroke'], y_pred)
error = 1 - metrics.accuracy_score(y_test['stroke'], y_pred)
precision = metrics.precision_score(y_test['stroke'], y_pred, average = None)
recall = metrics.recall_score(y_test['stroke'], y_pred, average = None)
F1_score = metrics.f1_score(y_test['stroke'], y_pred, average = None)
print([accuracy, error, precision, recall, F1_score])
```
